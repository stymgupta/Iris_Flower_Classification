## **Standardizing the data**

```python
from sklearn.preprocessing import StandardScaler
x=StandardScaler().fit_transform(x)
x
```

## **Dividing the Model into training and testing Model**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

## **Time to Train the Model**

```python
from sklearn import linear_model
```

```python
reg=linear_model.LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

## Training Finished...Model Testing Begins

## **To Check the accuracy of the testing model and the complete Model**

```python
reg.score(x_test, y_test)    #Accuracy Of the Testing Model
```

```python
reg.score(x,y)    #Accuracy of the Complete Model
```
